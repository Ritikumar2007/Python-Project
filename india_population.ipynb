```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = '/mnt/data/india_population.csv'
data = pd.read_csv(r"C:\Users\Mahir\Downloads\india_population.csv")
data.sort_values(by='Year', inplace=True)

# chart1: Population Growth Over the Years
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Population'], marker='o', color='b', label='Population')
plt.title('India Population Growth Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.grid(alpha=0.5)
plt.legend()
plt.show()

# Chart 2: Yearly % Change Over the Years
plt.figure(figsize=(10, 6))
plt.bar(data['Year'], data['Yearly % Change'], color='orange', alpha=0.7, label='Yearly % Change')
plt.title('Yearly Percentage Change in Population', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Yearly % Change (%)', fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.legend()
plt.show()

# Chart 3: Urban Population Percentage Over the Years
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Urban Pop %'], marker='o', color='g', label='Urban Population %')
plt.title('Urban Population Percentage Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Urban Population %', fontsize=12)
plt.grid(alpha=0.5)
plt.legend()
plt.show()

# Chart 4: Population Density Over the Years
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Density (P/Km²)'], marker='o', color='r', label='Population Density')
plt.title('Population Density Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Density (P/Km²)', fontsize=12)
plt.grid(alpha=0.5)
plt.legend()
plt.show()
```
